**Random Forest comes under Ensemble Learning which is basically combining multiple algorithms or using an algorithm multiple times to produce something more powerful than the original.**

# **STEPS OF RANDOM FOREST ALGORITHM:**

**STEP 1**: Pick at random K data points from the Training Set.

**STEP 2**: Build the Decision Tree associated to these K data points.

**STEP 3**: Choose the number Ntree of trees you want to build(usually 500) and repeat steps 1 & 2.

**STEP 4**: For a new data point, make each one of your Ntree trees predict the value of Y for the data point in question,and assign the new data point the average across all of the predicted Y values.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/content/Position_Salaries.csv')
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.ensemble import RandomForestRegressor

# Every tree is an estimator, here number of estimators/trees = 10
regressor = RandomForestRegressor(n_estimators = 10,random_state = 0)
regressor.fit(x,y)
```

```python
regressor.predict([[6.5]])
```

```python
x_grid = np.arange(min(x),max(x),0.1)
x_grid = x_grid.reshape(len(x_grid) , 1)
plt.scatter(x,y, color = 'red')
plt.plot(x_grid, regressor.predict(x_grid),color = 'blue')
plt.title('Truth or Bluff? (Random Forest Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salaries')
plt.show()
```
