**The dataset contains only 1 independent and dependent variable i.e its short. Decision tree algorithm will work on this dataset but it works more efficiently on big datasets. The code here will also work for any other dataset.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
print(dataset)
```

```python
x = dataset.iloc[:, 1:-1].values
print(x)
```

```python
y = dataset.iloc[:, -1].values
print(y)
```

**THERE IS NO NEED OF APPLYING FEATURE SCALING WHILE USING DECISION TREES OR RANDOM FOREST ALGORITHMS !!!!**

```python
from sklearn.tree import DecisionTreeRegressor
regressor =  DecisionTreeRegressor(random_state = 0)
regressor.fit(x,y)
```

```python
regressor.predict([[6.5]])
```

```python
y_pred = regressor.predict(x)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,y_pred)
```

```python
plt.scatter(x,y, color = 'red') # plotting the actual dataset
plt.plot(x, regressor.predict(x),color = 'blue') # plotting dataset with predicted salaries
plt.title('Truth or Bluff? (Decision Tree Regression)') # giving title to the visualization
plt.xlabel('Position Level')
plt.ylabel('Salaries')
plt.show()
```

```python
x_grid = np.arange(min(x),max(x),0.1)
x_grid = x_grid.reshape(len(x_grid) , 1)
plt.scatter(x,y, color = 'red')
plt.plot(x_grid, regressor.predict(x_grid),color = 'blue')
plt.title('Truth or Bluff? (Decision Tree Regression)') # giving title to the visualization
plt.xlabel('Position Level')
plt.ylabel('Salaries')
plt.show()
```
