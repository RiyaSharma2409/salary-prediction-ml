**WE ARE USING SVR ON THE SAME DATASET WE USED POLYNOMIAL REGRESSION TO COMPARE THE RESULTS. THIS MODEL CONTAINS FEATURE SCALING IN VERY DEEP.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/content/Position_Salaries.csv')
```

```python
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
print(x)
```

```python
print(y)
```

# **FEATURE SCALING**

we are gonna apply feature scaling to both x and y values before spliting the dataset because only then svr works.

```python
y = y.reshape(len(y),1)
```

```python
from sklearn.preprocessing import StandardScaler
ss_x = StandardScaler()
ss_y = StandardScaler()
x = ss_x.fit_transform(x)
y = ss_y.fit_transform(y)
print(x)
print(y)
```

```python
from sklearn.svm import SVR
regressor = SVR( kernel = 'rbf')
regressor.fit(x,y)
```

```python
ss_y.inverse_transform(regressor.predict(ss_x.transform([[6.5]])).reshape(-1,1))
```

```python
plt.scatter(ss_x.inverse_transform(x),ss_y.inverse_transform(y), color = 'red')
plt.plot(ss_x.inverse_transform(x),ss_y.inverse_transform(regressor.predict(x).reshape(-1,1)), color = 'blue')
plt.title('Truth or Bluff? (Support Vector Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salaries')
plt.show()
```

the data point at the top right corner is an outlier
